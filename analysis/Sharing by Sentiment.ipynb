```python
# let's "kind of" replicate the graph in the social deviance study
# first get quantiles based on RTs
rt = df.sort('rt',ascending=False).dropna()
```

```python
def quantile_list(df, quantiles):
     return type(df)({q: df.quantile(q) for q in quantiles})
     # this will work for both Series and DataFrames
        
quantile_list(rt, np.arange(0,1,0.2))
```

```python
# polarity values for quintiles averaged...
#let's get the quantiles for RTs
rts = quantile_list(rt, np.arange(0,1,0.2)).ix['rt'].values.tolist() + [rt.rt.max()]
y = []
for q in range(len(rts)-1):
    print(str('Q'+str(q+1)),': {0:7.2f} - {1:7.2f}'.format(rts[q],rts[q+1]), sep=' ',end='\t')
    quintile = df[(rts[q] <= df['rt']) & (df['rt'] < rts[q+1])]
    # qmean = quintile.spnews.mean()
    # weight = np.sum(quintile.spnews * quintile.rt)*100/quintile.rt.sum()
    median = quintile[quintile.spnews > df.spnews.median()].rt.sum() *100 / quintile.rt.sum()
    # mean =  quintile[quintile.spnews > df.spnews.mean()].rt.sum() *100 / quintile.rt.sum()
    y.append(median)
    
    print('RT weighted sentiment polarity of news:{0:5.2f}'.format(y[-1]))
```

```python
plt.barh(range(5),y)
plt.xlabel('Sentiment Polarity')
plt.ylabel('RT Quintiles (Q5 is top, Q1 is botttom)')
plt.title('Mean of polarities accross quintiles')
plt.yticks(np.arange(5) + 0.35, ('Q1', 'Q2', 'Q3', 'Q4', 'Q5'));
```

```python
#let's replicate the graph in the social deviance study
#firt get quantiles based on RTs
rt = df.sort('rt',ascending=False).dropna()
```

```python
def quantile_list(df, quantiles):
     return type(df)({q: df.quantile(q) for q in quantiles})
     # this will work for both Series and DataFrames

quantile_list(df, np.arange(0.1,1,0.1))
```

```python
rts = [0] + quantile_list(df, np.arange(0.1,1,0.1)).ix['rt'].values.tolist()
rts
```

```python
#quantiles averaged, will look into this...
for q in range(len(rts)-1):
    print(rts[q],rts[q+1],sep='\t',end='\t')
    print(df[(rts[q]<df['rt'])&(df['rt']<rts[q+1])].polarity.mean())
```

# Descriptive Statistics of Tweets

This notebook is solely based on Twitter data of [CNN](twitter.com/cnn) and [NYTIMES](twitter.com/nytimes). Downloaded the JSON files containing last 3200 status messages sent by these accounts using [DD-CSS](dd-css.com) and converted them to CSV files (X-tweets.csv) using the script in scrapers/parse_tweets.py

Description of the CNN-tweets.csv and NYT-tweets.csv files:
* **twid, tweep, twtext, fav, rt, url**
* twid: id of the nytimes' tweet (note: in case of an RT, the original tweet has a different id)
* tweep: if RTed, then the screen name of the original tweep, otherwise nytimes.
* twtext: text of the tweet. Note:if RTed, then the text of the original tweet.
* fav: favorites count.
* rt: retweet count.
* url: expanded form of the url(s) in the tweet. if no url, then empty. if multiple urls, then separated by a space character.

Some issues observed:
* Tweets with multiple URLS
* URLS that 'nyt' or 'cnn' is not a part of expanded URL

Then, using the news content, we append the following columns to our table:
* **dt, title, content**

# NYTIMES

```python
#read in nytimes data
df = pd.read_csv('data/NYT-tweets.csv',na_filter=False)
df
```

```python
# Tweets with multiple URLS
multip = df[df.url.str.contains(' ')].url
multip
```

```python
# URLS that 'nyt' is not a part of expanded URL
single = df[~df.url.str.contains('[\s]')].url
single = single[single.str.contains('http')]
single[~single.str.contains('nyt')]
```

```python
# Number of tweets with single URL
len(single)
```

```python
# Number of tweets with single URL containing NYT in its URL
len(single[single.str.contains('nyt|NYT')])
```

#CNN

```python
#read in CNN data
df = pd.read_csv('data/CNN-tweets.csv',na_filter=False)
df
```

```python
# Tweets with multiple URLS
multip = df[df.url.str.contains(' ')].url
multip
```

```python
# URLS that 'CNN' is not a part of expanded URL
single = df[~df.url.str.contains('[\s]')].url
single = single[single.str.contains('http')]
single[~single.str.contains('cnn|CNN')]
```

```python
# Number of tweets with single URL
len(single)
```

```python
# Number of tweets with single URL containing CNN in its URL
len(single[single.str.contains('cnn|CNN')])
```

```python
cnn = df
```

```python
nyt._metadata = 'NYT'
cnn._metadata = 'CNN'
fields = ['url0','url1','url2']
res = pd.DataFrame(index=fields)
field = {}
for df in (nyt,cnn):
    field['url0'] = len(df[df.url == ''])
    field['url1'] = len(df[df.url.str.contains('http') & ~df.url.str.contains('[\s]')])
    field['url2'] = len(df[df.url.str.contains(' ')].url)
    res = res.join(pd.DataFrame(data=[field[f] for f in fields],index=fields,columns=[df._metadata]))
res
```
